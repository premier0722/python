```python
from selenium import webdriver #操作網頁
import time #延遲時間操作
import csv
from bs4 import BeautifulSoup #倒入網頁資料
from selenium.webdriver.common.keys import Keys #操作鍵盤使用

buy=str(input("請輸入要搜尋的商品："))

url1='https://www.momoshop.com.tw/main/Main.jsp'
chrome_path='C:\executable_path\chromedriver.exe'
driver = webdriver.Chrome(chrome_path)
driver.get(url1)
driver.maximize_window()

element=driver.find_element_by_id('keyword')
element.send_keys(buy)
element.send_keys(Keys.ENTER)

soup = BeautifulSoup(driver.page_source, 'html5lib')
#print(soup)

sels=soup.find('div','listArea').find_all('li')
#print(sels)

with open('output.csv','w',newline='') as csvfile:
    writer=csv.writer(csvfile)
    writer.writerow(['momo購物台'])

for sel in sels:
    name=sel.find('p','prdName').text.strip()
    price=sel.find('span','price').text.strip()
    print(name,price)
    
    with open('output.csv','a',newline='') as csvfile:
        writer=csv.writer(csvfile)
        writer.writerow([name,price])
```
